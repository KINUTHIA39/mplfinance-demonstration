```python
import yfinance as yf 
import pandas as pd 
from tiingo import TiingoClient 
import mplfinance as mpf 
import os 
import datetime 

```

```python
def fetch_yahoo_data(symbol, start_date_str, end_date_str):
    interval = '1d'
    data = yf.download(tickers = 'ETH-USD', start = start_date_str, end = end_date_str, interval = '1d')
    price_df = pd.DataFrame(data)
    return price_df


symbol = 'ETH-USD'
now = datetime.datetime.now()
end_date_str = now.strftime("%Y-%m-%d")
start_date = now - datetime.timedelta(days=120)
start_date_str = start_date.strftime("%Y-%m-%d")

df = fetch_yahoo_data(symbol, start_date_str, end_date_str)
```
